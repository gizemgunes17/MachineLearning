```python
def plotLearningCurve(est, X_train, y_train, X_test, y_test, iterNum=100, isReg=True, y_lower=0.2, y_upper=1):
  x_values = []
  y_values_train = []
  y_values_test = []

  items = np.linspace(100, len(X_train), iterNum)
  for i in items:
    i = int(i)

    # sample train data
    X_train_sample = X_train[:i]
    y_train_sample = y_train[:i]

    # create model
    est.fit(X_train_sample, y_train_sample)

    # train evals
    y_train_pred = est.predict(X_train_sample)
    if isReg:
      train_score = mean_squared_error(y_train_sample, y_train_pred, squared=False)
    else:
      train_score = f1_score(y_train_sample, y_train_pred,average='weighted')

    # test evals
    y_test_pred = est.predict(X_test)
    if isReg:
      test_score =mean_squared_error(y_test, y_test_pred, squared=False)
    else:
      test_score = f1_score(y_test, y_test_pred, average='weighted')

    # fill the lists
    x_values.append(i)
    y_values_train.append(train_score)
    y_values_test.append(test_score)
  
  plt.figure(figsize=(10, 6))
  plt.plot(x_values, y_values_train, label="train")
  plt.plot(x_values, y_values_test, label="test")
  plt.legend()
  plt.grid(True)
  plt.ylim(y_lower, y_upper)
  plt.show()
  
```
